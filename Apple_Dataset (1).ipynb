```python
x = data_1.head(7)['Product Name']
y = data_1.head(7)['Discount Percentage']
plt.pie(y, labels = x, radius = 3, autopct='%1.1f%%')
plt.legend(loc="best")
plt.xticks(rotation=30)
plt.title('Discounts offered on iPhones')
plt.show()
```
